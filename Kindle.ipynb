```python
class Book:
    def __init__(self, book_id, title, published_date, price, author, genre):
        self.book_id = book_id
        self.title = title
        self.published_date = published_date
        self.price = price
        self.author = author
        self.genre = genre

class Review:
    def __init__(self, review_id, book, rating, review_text, review_date):
        self.review_id = review_id
        self.book = book
        self.rating = rating
        self.review_text = review_text
        self.review_date = review_date

class Author:
    def __init__(self, author_id, first_name, last_name, bio, genre):
        self.author_id = author_id
        self.first_name = first_name
        self.last_name = last_name
        self.bio = bio
        self.genre = genre

class Genre:
    def __init__(self, genre_id, genre_type):
        self.genre_id = genre_id
        self.genre_type = genre_type

class AmazonKindle:
    def __init__(self):
        self.books = []
        self.reviews = []
        self.authors = []
        self.genres = []

    def add_book(self, book):
        self.books.append(book)

    def add_review(self, review):
        self.reviews.append(review)

    def add_author(self, author):
        self.authors.append(author)

    def add_genre(self, genre):
        self.genres.append(genre)

    def get_books_by_author(self, author_id):
        return [book for book in self.books if book.author.author_id == author_id]

    def get_books_by_genre(self, genre_id):
        return [book for book in self.books if book.genre.genre_id == genre_id]

    def get_average_rating(self, book_id):
        reviews = [review for review in self.reviews if review.book.book_id == book_id]
        if reviews:
            return sum(review.rating for review in reviews) / len(reviews)
        else:
            return 0
```
