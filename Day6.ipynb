```python
# IndexError
lst = [1, 2, 3]
print(lst[5])
```

```python
# Handle IndexError
try:
    lst = [1, 2, 3]
    print(lst[5])
except IndexError as e:
    print(f"IndexError: {e}")
```

```python
# AssertionError
assert 1 == 0, "This is an AssertionError"
```

```python
# Handle AssertionError
try:
    assert 1 == 0, "This is an AssertionError"
except AssertionError as e:
    print(f"AssertionError: {e}")
```

```python
#AttributeError
x = 10
x.append(5)
```

```python
# Handle AttributeError
try:
    x = 10
    x.append(5)
except AttributeError as e:
    print(f"AttributeError: {e}")
```

```python
# ImportError
import non_existent_module
```

```python
# Handle ImportError
try:
    import non_existent_module
except ImportError as e:
    print(f"ImportError: {e}")
```

```python
# KeyError
d = {"a": 1}
print(d["b"])
```

```python
# Handle KeyError
try:
    d = {"a": 1}
    print(d["b"])
except KeyError as e:
    print(f"KeyError: {e}")
```

```python
#  NameError
print(non_existent_variable)
```

```python
# Handle NameError
try:
    print(non_existent_variable)
except NameError as e:
    print(f"NameError: {e}")
```

```python
# MemoryError
a = []
while True:
    a.append(' ' * 10**6)
```

```python
# Handle MemoryError
try:
    a = []
    while True:
        a.append(' ' * 10**6)
except MemoryError as e:
    print(f"MemoryError: {e}")
```

```python
# TypeError
print("5" + 5)
```

```python
# Handle TypeError
try:
    print("5" + 5)
except TypeError as e:
    print(f"TypeError: {e}")
```

```python
import pandas as pd
import numpy as np

#  Missing Values Error
df = pd.DataFrame({
    "A": [1, 2, np.nan, 4],
    "B": [np.nan, 2, 3, 4]
})

print(df)
```

```python
# Handle Missing Values
df.fillna(0, inplace=True)
print(df)
```

```python
# Incorrect Data Types Error
df = pd.DataFrame({
    "A": ["1", "2", "three", "4"]
})

print(df)
```

```python
# Handle Incorrect Data Types
df["A"] = pd.to_numeric(df["A"], errors='coerce')
print(df)
```

```python
# Outliers Error
data = [1, 2, 3, 4, 1000]
print(data)
```

```python
# Handle Outliers
import numpy as np

def remove_outliers(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return [x for x in data if lower_bound <= x <= upper_bound]

cleaned_data = remove_outliers(data)
print(cleaned_data)
```

```python
# Reproduce Incorrect Formulas Error
df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})
df["C"] = df["A"] / df["B"]
print(df)
```

```python
# Handle -Incorrect Formulas
df["C"] = df["A"] * df["B"]
print(df)
```
